```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)

purchase_data.head()
```

## Player Count

* Display the total number of players


```python
purchase_describe = len(purchase_data['SN'].unique())

print("Player Count = " + str(purchase_describe))
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
Number_Of_Unique_Items = len(purchase_data['Item ID'].value_counts())
Average_Price = purchase_data['Price'].mean(),
Number_Of_Purchases = len(purchase_data['Purchase ID'].unique()),
Total_Revenue = purchase_data['Price'].sum()

x = pd.Series("")

Purchasing_Analysis_df= pd.DataFrame(x)
Purchasing_Analysis_df['Number Of Unique Items']= Number_Of_Unique_Items
Purchasing_Analysis_df['Average Price']= Average_Price
Purchasing_Analysis_df['Number Of Purchases']= Number_Of_Purchases
Purchasing_Analysis_df['Total Revenue']= Total_Revenue

Purchasing_Analysis_df_revised = Purchasing_Analysis_df[['Number Of Unique Items','Average Price', 'Number Of Purchases', 'Total Revenue']]
Purchasing_Analysis_df_revised
Purchasing_Analysis_df_revised['Average Price'] = Purchasing_Analysis_df_revised['Average Price'].map("${:.2f}".format)
Purchasing_Analysis_df_revised['Total Revenue'] = Purchasing_Analysis_df_revised['Total Revenue'].map("${:,.2f}".format)
Purchasing_Analysis_df_revised
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
max_age = purchase_data['Age'].max()
bins = [0,9,14,19,24,29,34,39,max_age]
labels = ['<10','10-14','15-19','20-24','25-29','30-34','35-39','40+']
purchase_data_df['Age Group'] = pd.cut(purchase_data_df['Age'], bins = bins, labels = labels, include_lowest = True)
purchase_data_df
Age_Group = purchase_data_df.groupby('Age Group')

Player_Count_A= Age_Group['Purchase ID'].count()
POPA= Player_Count_A / Player_Count_A.sum() 

age_group_df = pd.DataFrame({"Player Count": Player_Count_A,
                             "Percentage of Total": POPA})

age_group_df['Percentage of Total'] = age_group_df['Percentage of Total'].map("{:.2%}".format)

age_group_df
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data['Age Group'] = pd.cut(purchase_data['Age'], bins = bins, labels = labels, include_lowest = True)
Age_Group_All_Pur = purchase_data.groupby('Age Group')
Age_Group_All_Pur.count()
Purchase_Count_A = Age_Group_All_Pur['Purchase ID'].count()
Total_Purchase_Value_A = Age_Group_All_Pur['Price'].sum()
Average_Purchase_Price_A = Total_Purchase_Value_A / Purchase_Count_A
Average_Total_Purchase_Per_Person_A = Total_Purchase_Value_A / Player_Count_A

age_group_df_dos = pd.DataFrame({'Purchase Count': Purchase_Count_A,
                                 'Average Purchase Price': Average_Purchase_Price_A,
                                 'Total Purchase Value': Total_Purchase_Value_A,
                                 'Average Total Purchase Per Person': Average_Total_Purchase_Per_Person_A})

age_group_df_dos['Average Purchase Price'] = age_group_df_dos['Average Purchase Price'].map("${:,.2f}".format)
age_group_df_dos['Total Purchase Value'] = age_group_df_dos['Total Purchase Value'].map("${:,.2f}".format)
age_group_df_dos['Average Total Purchase Per Person'] = age_group_df_dos['Average Total Purchase Per Person'].map("${:,.2f}".format)

age_group_df_dos
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
SN_Group = purchase_data.groupby('SN')

Purchase_Count_By_SN = SN_Group['Purchase ID'].count()
Total_Purchase_Value_SN = SN_Group['Price'].sum()
Average_Purchase_Price_SN = Total_Purchase_Value_SN / Purchase_Count_By_SN

spender_df = pd.DataFrame({"Purchase Count": Purchase_Count_By_SN,
                           "Average Purchase Price": Average_Purchase_Price_SN,
                           "Total Purchase Value": Total_Purchase_Value_SN})

#spender_df['Average Purchase Price'] = spender_df['Average Purchase Price'].map("${:,.2f}".format)
#spender_df['Total Purchase Value'] = spender_df['Total Purchase Value'].map("${:,.2f}".format)

spender_df_descending = spender_df.sort_values("Total Purchase Value", ascending = False)


spender_df_descending['Average Purchase Price'] = spender_df_descending['Average Purchase Price'].map("${:,.2f}".format)
spender_df_descending['Total Purchase Value'] = spender_df_descending['Total Purchase Value'].map("${:,.2f}".format)
spender_df_descending.head()

```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, average item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
#Item_Data_unique = purchase_data.drop_duplicates(subset=['Item Name'])
#Item_Data = purchase_data.groupby('Item ')

Item_Group = purchase_data.groupby(['Item ID','Item Name'])
#Item_Group_2 = Item_Group.groupby('Item Name')

#Item_Name = Item_Group['Item Name'].unique()
Item_Purchase_Count = Item_Group['Purchase ID'].count()

Item_Price = Item_Group['Price'].mean()
Item_Total_Purchase_Value = Item_Price * Item_Purchase_Count

item_df = pd.DataFrame({#"Item Name" : Item_Name,
                        "Purchase Count": Item_Purchase_Count,
                        "Item Price": Item_Price,
                        "Total Purchase Value": Item_Total_Purchase_Value})

item_df['Item Price'] = item_df['Item Price'].map("${:,.2f}".format)
item_df['Total Purchase Value'] = item_df['Total Purchase Value'].map("${:.2f}".format)

item_df_descending = item_df.sort_values("Purchase Count", ascending = False)

item_df_descending.head()




```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_df_tpp_desc = pd.DataFrame({#"Item Name" : Item_Name,
                        "Purchase Count": Item_Purchase_Count,
                        "Item Price": Item_Price,
                        "Total Purchase Value": Item_Total_Purchase_Value})
item_df_tpp_desc2 = item_df_tpp_desc.sort_values("Total Purchase Value", ascending = False)

item_df_tpp_desc2['Item Price'] = item_df_tpp_desc2['Item Price'].map("${:,.2f}".format)
item_df_tpp_desc2['Total Purchase Value'] = item_df_tpp_desc2['Total Purchase Value'].map("${:.2f}".format)

item_df_tpp_desc2.head()

```
